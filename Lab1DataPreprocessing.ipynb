```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
clean_data=pd.read_csv('ecoli.csv')
sb.pairplot(clean_data,hue='Class_Distribution')
```

```python
clean_data.median()
```

```python
clean_data.mode()
```

```python
plt.figure(figsize=(15, 15))
for column_index, column in enumerate(clean_data.columns):
    if column == 'Class_Distribution':
        continue
    plt.subplot(3, 3, column_index + 1)
    sb.violinplot(x='Class_Distribution', y=column, data=clean_data)
plt.savefig('./everything2.png', dpi=300)
```

```python
plt.figure(figsize=(15, 15))
for column_index, column in enumerate(clean_data.columns):
    if column == 'Class_Distribution':
        continue
    plt.subplot(3, 3, column_index + 1)
    sb.boxplot(x='Class_Distribution', y=column, data=clean_data)
plt.savefig('./everything3.png', dpi=300)
```
